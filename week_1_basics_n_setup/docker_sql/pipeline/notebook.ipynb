### Optimized Data Ingestion
<p>We won't insert data all at once. Do it in batches and use an iterator(Chuncked reading)</p>

```python
# We split this entire data in chunks and insert one by one
len(df)
```

```python
# Creating the iterator
df_iter = pd.read_csv(
    url,
    dtype=dtype,
    parse_dates=parse_dates,
    iterator=True,
    chunksize=100000,
)
```

```python
# Iterating over chunks
for df_chunk in df_iter:
    print(len(df_chunk))
```

```python
# from arabic 'taqaddum' means progress, tqdm is a library to visualize progress bar on loops and iterables
from tqdm.auto import tqdm 
```

```python
# Iterate over the chunks and insert on the specified database
for df_chunk in tqdm(df_iter):
    df_chunk.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')
```
