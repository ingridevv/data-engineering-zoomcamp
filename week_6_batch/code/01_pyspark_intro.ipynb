```python
import pandas as pd
import pyspark
from pyspark.sql import SparkSession
```

```python
# Initialize spark session
spark = SparkSession.builder \
    .master("local[*]") \
    .appName('spark_intro') \
    .getOrCreate()
```

```python
# Read data saved in-memory
df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv("data/fhvhv_tripdata_2021-01.csv.gz")
```
